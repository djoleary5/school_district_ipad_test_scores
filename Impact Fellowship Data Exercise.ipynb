```python
# Dennis O'Leary
# Dependencies 
import pandas as pd
```

```python
#raw ELA data
ela_data = pd.read_csv("ela.csv")
ela_data.head()
```

```python
#raw Math data
math_data = pd.read_csv("math.csv")
math_data.head()
```

```python
#raw Demographic data
demo_data = pd.read_csv("demo.csv")
demo_data.head()
```

```python
# filter for ELA/Literacy Grade 5 test in 2017 for pilot schools (610534 609880 610066 610082 610182) 

pilot_ela_5_2017 = ela_data.loc[((ela_data['School ID'] == 610534) |
                         (ela_data['School ID'] == 609880) |
                         (ela_data['School ID'] == 610066) |
                         (ela_data['School ID'] == 610082) |
                         (ela_data['School ID'] == 610182)) &
                        (ela_data['Test Name'] == 'ELA/Literacy Grade 5') &
                        (ela_data['Year'] == 2017)]
pilot_ela_5_2017['Overall ELA Mean Scale Score'] = pilot_ela_5_2017['Overall ELA Mean Scale Score'].astype(int)
pilot_ela_5_2017['# Students Tested'] = pilot_ela_5_2017['# Students Tested'].astype(int)
pilot_ela_5_2017['Overall ELA Mean Scale Score Weighted'] = pilot_ela_5_2017['Overall ELA Mean Scale Score']*pilot_ela_5_2017['# Students Tested']                    
pilot_ela_5_2017 = pilot_ela_5_2017.drop_duplicates()
pilot_ela_5_2017
```

```python
# calculate the average 2017 ELA/Literacy Grade 5 test score of all the pilot schools combined

pilot_ela_5_2017_students_tested = list(map(int, pilot_ela_5_2017['# Students Tested']))
pilot_ela_5_2017_students_tested_total = sum(pilot_ela_5_2017_students_tested)
print("Number of students tested for 2017 ELA/Literacy Grade 5 in pilot schools: " + str(pilot_ela_5_2017_students_tested_total))

pilot_ela_5_2017_overall_mean_weighted = list(map(int, pilot_ela_5_2017['Overall ELA Mean Scale Score Weighted']))
pilot_ela_5_2017_overall_mean_total = sum(pilot_ela_5_2017_overall_mean_weighted)

pilot_ela_5_2017_avg_score = pilot_ela_5_2017_overall_mean_total/pilot_ela_5_2017_students_tested_total

print("Average score of all the pilot schools combined: " + str(round(pilot_ela_5_2017_avg_score, 0)))
```

```python
# filter for ELA/Literacy Grade 5 test in 2017 for nonpilot schools

nonpilot_ela_5_2017 = ela_data.loc[(ela_data['School ID'] != 610534) &
                         (ela_data['School ID'] != 609880) &
                         (ela_data['School ID'] != 610066) &
                         (ela_data['School ID'] != 610082) &
                         (ela_data['School ID'] != 610182) &
                        (ela_data['Test Name'] == 'ELA/Literacy Grade 5') &
                        (ela_data['Year'] == 2017)]
nonpilot_ela_5_2017['Overall ELA Mean Scale Score'] = nonpilot_ela_5_2017['Overall ELA Mean Scale Score'].astype(int)
nonpilot_ela_5_2017['# Students Tested'] = nonpilot_ela_5_2017['# Students Tested'].astype(int)
nonpilot_ela_5_2017['Overall ELA Mean Scale Score Weighted'] = nonpilot_ela_5_2017['Overall ELA Mean Scale Score']*nonpilot_ela_5_2017['# Students Tested']                    
nonpilot_ela_5_2017 = nonpilot_ela_5_2017.drop_duplicates()
nonpilot_ela_5_2017.head()
```

```python
# calculate the average 2017 ELA/Literacy Grade 5 test score of all the nonpilot schools combined

nonpilot_ela_5_2017_students_tested = list(map(int, nonpilot_ela_5_2017['# Students Tested']))
nonpilot_ela_5_2017_students_tested_total = sum(nonpilot_ela_5_2017_students_tested)
print("Number of students tested for 2017 ELA/Literacy Grade 5 in nonpilot schools: " + str(nonpilot_ela_5_2017_students_tested_total))

nonpilot_ela_5_2017_overall_mean_weighted = list(map(int, nonpilot_ela_5_2017['Overall ELA Mean Scale Score Weighted']))
nonpilot_ela_5_2017_overall_mean_total = sum(nonpilot_ela_5_2017_overall_mean_weighted)

nonpilot_ela_5_2017_avg_score = nonpilot_ela_5_2017_overall_mean_total/nonpilot_ela_5_2017_students_tested_total

print("Average score of all the nonpilot schools combined: " + str(round(nonpilot_ela_5_2017_avg_score, 0)))
```

```python
# filter for all ELA tests in 2016 for pilot schools (610534 609880 610066 610082 610182) 

pilot_ela_2016 = ela_data.loc[((ela_data['School ID'] == 610534) |
                         (ela_data['School ID'] == 609880) |
                         (ela_data['School ID'] == 610066) |
                         (ela_data['School ID'] == 610082) |
                         (ela_data['School ID'] == 610182)) &
                        (ela_data['Year'] == 2016)  &
                        (ela_data['Overall ELA Mean Scale Score'] != " ")]
pilot_ela_2016['Overall ELA Mean Scale Score'] = pilot_ela_2016['Overall ELA Mean Scale Score'].astype(int)
pilot_ela_2016['# Students Tested'] = pilot_ela_2016['# Students Tested'].astype(int)
pilot_ela_2016['Overall ELA Mean Scale Score Weighted'] = pilot_ela_2016['Overall ELA Mean Scale Score']*pilot_ela_2016['# Students Tested']                    
pilot_ela_2016 = pilot_ela_2016.drop_duplicates()
pilot_ela_2016
```

```python
# calculate the 2016 Average Scores for each pilot school

pilot_ela_2016_grp = pilot_ela_2016.groupby(pilot_ela_2016['School ID']).sum()
del pilot_ela_2016_grp['Year']
del pilot_ela_2016_grp['Overall ELA Mean Scale Score']
pilot_ela_2016_grp['2016 Average Score'] = pilot_ela_2016_grp['Overall ELA Mean Scale Score Weighted']/pilot_ela_2016_grp['# Students Tested']
pilot_ela_2016_grp.sort_values(by=['2016 Average Score'], inplace=True, ascending=False)
pilot_ela_2016_grp
```

```python
# filter for all ELA tests in 2017 for pilot schools (610534 609880 610066 610082 610182) 

pilot_ela_2017 = ela_data.loc[((ela_data['School ID'] == 610534) |
                         (ela_data['School ID'] == 609880) |
                         (ela_data['School ID'] == 610066) |
                         (ela_data['School ID'] == 610082) |
                         (ela_data['School ID'] == 610182)) &
                        (ela_data['Year'] == 2017)  &
                        (ela_data['Overall ELA Mean Scale Score'] != " ")]
pilot_ela_2017['Overall ELA Mean Scale Score'] = pilot_ela_2017['Overall ELA Mean Scale Score'].astype(int)
pilot_ela_2017['# Students Tested'] = pilot_ela_2017['# Students Tested'].astype(int)
pilot_ela_2017['Overall ELA Mean Scale Score Weighted'] = pilot_ela_2017['Overall ELA Mean Scale Score']*pilot_ela_2017['# Students Tested']                    
pilot_ela_2017 = pilot_ela_2017.drop_duplicates()
pilot_ela_2017.head()
```

```python
# calculate the 2017 Average Scores for each pilot school

pilot_ela_2017_grp = pilot_ela_2017.groupby(pilot_ela_2017['School ID']).sum()
del pilot_ela_2017_grp['Year']
del pilot_ela_2017_grp['Overall ELA Mean Scale Score']
pilot_ela_2017_grp['2017 Average Score'] = pilot_ela_2017_grp['Overall ELA Mean Scale Score Weighted']/pilot_ela_2017_grp['# Students Tested']
pilot_ela_2017_grp.sort_values(by=['2017 Average Score'], inplace=True, ascending=False)
pilot_ela_2017_grp
```

```python
# calculate the change in Average Scores for each pilot school from 2016 to 2017

pilot_ela_2016_2017 = pd.merge(pilot_ela_2016_grp, pilot_ela_2017_grp, on='School ID', how='outer')
del pilot_ela_2016_2017['# Students Tested_x']
del pilot_ela_2016_2017['# Students Tested_y']
del pilot_ela_2016_2017['Overall ELA Mean Scale Score Weighted_x']
del pilot_ela_2016_2017['Overall ELA Mean Scale Score Weighted_y']
pilot_ela_2016_2017['Change in Average Score'] = pilot_ela_2016_2017['2017 Average Score'] - pilot_ela_2016_2017['2016 Average Score']
pilot_ela_2016_2017.sort_values(by=['Change in Average Score'], inplace=True, ascending=False)
pilot_ela_2016_2017
```

```python
# filter for ELA/Literacy Grade 5 test in 2016 for pilot schools (610534 609880 610066 610082 610182) 

pilot_ela_5_2016 = ela_data.loc[((ela_data['School ID'] == 610534) |
                         (ela_data['School ID'] == 609880) |
                         (ela_data['School ID'] == 610066) |
                         (ela_data['School ID'] == 610082) |
                         (ela_data['School ID'] == 610182)) &
                        (ela_data['Year'] == 2016)  &
                        (ela_data['Test Name'] == 'ELA/Literacy Grade 5')]
pilot_ela_5_2016['2016 Overall ELA Mean Scale Score'] = pilot_ela_5_2016['Overall ELA Mean Scale Score'].astype(int)
pilot_ela_5_2016['2016 # Students Tested'] = pilot_ela_5_2016['# Students Tested'].astype(int)
pilot_ela_5_2016['2016 Overall ELA Mean Scale Score Weighted'] = pilot_ela_5_2016['2016 Overall ELA Mean Scale Score']*pilot_ela_5_2016['2016 # Students Tested'] 
pilot_ela_5_2016 = pilot_ela_5_2016[['School ID', '2016 # Students Tested', '2016 Overall ELA Mean Scale Score', '2016 Overall ELA Mean Scale Score Weighted']]
pilot_ela_5_2016 = pilot_ela_5_2016.drop_duplicates()
pilot_ela_5_2016
```

```python
# filter for ELA/Literacy Grade 5 test in 2017 for pilot schools (610534 609880 610066 610082 610182) 

pilot_ela_5_2017 = ela_data.loc[((ela_data['School ID'] == 610534) |
                         (ela_data['School ID'] == 609880) |
                         (ela_data['School ID'] == 610066) |
                         (ela_data['School ID'] == 610082) |
                         (ela_data['School ID'] == 610182)) &
                        (ela_data['Year'] == 2017)  &
                        (ela_data['Test Name'] == 'ELA/Literacy Grade 5')]
pilot_ela_5_2017['2017 Overall ELA Mean Scale Score'] = pilot_ela_5_2017['Overall ELA Mean Scale Score'].astype(int)
pilot_ela_5_2017['2017 # Students Tested'] = pilot_ela_5_2017['# Students Tested'].astype(int)
pilot_ela_5_2017['2017 Overall ELA Mean Scale Score Weighted'] = pilot_ela_5_2017['2017 Overall ELA Mean Scale Score']*pilot_ela_5_2017['2017 # Students Tested'] 
pilot_ela_5_2017 = pilot_ela_5_2017[['School ID', '2017 # Students Tested', '2017 Overall ELA Mean Scale Score', '2017 Overall ELA Mean Scale Score Weighted']]
pilot_ela_5_2017 = pilot_ela_5_2017.drop_duplicates()
pilot_ela_5_2017
```

```python
# calculate the change in average scores for ELA/Literacy Grade 5 for each pilot school from 2016 to 2017
# pilot school 610066 had the greatest increase in average score (14 points, 1.8%)

pilot_ela_5_2016_2017 = pd.merge(pilot_ela_5_2016, pilot_ela_5_2017, on='School ID', how='outer')
pilot_ela_5_2016_2017['Change(points)'] = pilot_ela_5_2016_2017['2017 Overall ELA Mean Scale Score'] - pilot_ela_5_2016_2017['2016 Overall ELA Mean Scale Score']
pilot_ela_5_2016_2017['Change(percent)'] = pilot_ela_5_2016_2017['Change(points)']/pilot_ela_5_2016_2017['2016 Overall ELA Mean Scale Score']
pilot_ela_5_2016_2017.sort_values(by=['Change(points)'], inplace=True, ascending=False)
pilot_ela_5_2016_2017
```

```python
# drop top pilot school and calculate overall average change for the other four pilot schools

pilot_ela_5_2016_2017 = pilot_ela_5_2016_2017[pilot_ela_5_2016_2017['School ID'] != 610066]
weighted_total_2016 = pilot_ela_5_2016_2017['2016 Overall ELA Mean Scale Score Weighted'].sum()
weighted_total_2017 = pilot_ela_5_2016_2017['2017 Overall ELA Mean Scale Score Weighted'].sum()
stu_tested_total_2016 = pilot_ela_5_2016_2017['2016 # Students Tested'].sum()
stu_tested_total_2017 = pilot_ela_5_2016_2017['2017 # Students Tested'].sum()
avg_2016 = weighted_total_2016/stu_tested_total_2016
avg_2017 = weighted_total_2017/stu_tested_total_2017
print('Average change(points) for remaining pilot schools: ' + str(avg_2017-avg_2016))
print('Average change(percent) for remaining pilot schools: ' + str((avg_2017-avg_2016)/avg_2016))
pilot_ela_5_2016_2017
```

```python
# filter for pilot schools and year 2017 from demographic data

pilot_demo_2017 = demo_data.loc[((demo_data['School ID'] == 610534) |
                         (demo_data['School ID'] == 609880) |
                         (demo_data['School ID'] == 610066) |
                         (demo_data['School ID'] == 610082) |
                         (demo_data['School ID'] == 610182)) &
                            (demo_data['School Year'] == "2017")]
pilot_demo_2017['Total'] = pilot_demo_2017['Total'].astype(int)
pilot_demo_2017['# Free/Reduced Lunch'] = pilot_demo_2017['# Free/Reduced Lunch'].astype(int)
pilot_stu_total = pilot_demo_2017['Total'].sum()
pilot_frl_total = pilot_demo_2017['# Free/Reduced Lunch'].sum()
pilot_frl_pct = pilot_frl_total/pilot_stu_total
print("Average percentage of students who received free/reduced lunch in the 5 pilot schools in 2017:" + str(pilot_frl_pct))
pilot_demo_2017
```

```python
# merge demographic data and non-pilot grade 5 2017 data with free lunch % greater than 22%

nonpilot_ela_demo_5_2017 = pd.merge(ela_data, demo_data, on='School ID', how='outer')
nonpilot_ela_demo_5_2017 = nonpilot_ela_demo_5_2017.loc[(nonpilot_ela_demo_5_2017['School ID'] != 610534) &
                         (nonpilot_ela_demo_5_2017['School ID'] != 609880) &
                         (nonpilot_ela_demo_5_2017['School ID'] != 610066) &
                         (nonpilot_ela_demo_5_2017['School ID'] != 610082) &
                         (nonpilot_ela_demo_5_2017['School ID'] != 610182) &
                        (nonpilot_ela_demo_5_2017['Test Name'] == 'ELA/Literacy Grade 5') &
                        (nonpilot_ela_demo_5_2017['Year'] == 2017) &
                        (nonpilot_ela_demo_5_2017['School Year'] == '2017') &
                        (nonpilot_ela_demo_5_2017['% Free/Reduced Lunch'] >= 0.22) ]
nonpilot_ela_demo_5_2017['Overall ELA Mean Scale Score'] = nonpilot_ela_demo_5_2017['Overall ELA Mean Scale Score'].astype(int)
nonpilot_ela_demo_5_2017['# Students Tested'] = nonpilot_ela_demo_5_2017['# Students Tested'].astype(int)
nonpilot_ela_demo_5_2017['Overall ELA Mean Scale Score Weighted'] = nonpilot_ela_demo_5_2017['Overall ELA Mean Scale Score']*nonpilot_ela_demo_5_2017['# Students Tested']
nonpilot_ela_demo_5_2017 = nonpilot_ela_demo_5_2017.drop_duplicates()


nonpilot_weighted_total = nonpilot_ela_demo_5_2017['Overall ELA Mean Scale Score Weighted'].sum()
nonpilot_stu_tested_total = nonpilot_ela_demo_5_2017['# Students Tested'].sum()
nonpilot_overall_avg = nonpilot_weighted_total/nonpilot_stu_tested_total
print("Average ELA mean test score of the non-pilot schools that have an equal or greater percentage of students who qualify for free/reduced lunch as the pilot schools:" + str(nonpilot_overall_avg))


nonpilot_ela_demo_5_2017
```
